# Cluster Analysis of the Piedmont Wine Dataset using k-Means
In this example, we perform clustering on the Piedmont Wine data set available at:

https://archive.ics.uci.edu/ml/machine-learning-databases/wine/

We begin by importing the needed libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import datasets
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from scipy.spatial.distance import cdist, pdist

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

np.set_printoptions(precision=5, suppress=True)  # suppress scientific float notation
```

We load the Piedmont wine data set and print some statistics.

```python
dataset = pd.read_csv("PiedmontWine.csv")
```

```python
dataset.describe()
```

```python
class_distribution = dataset['Class'].value_counts()
print(class_distribution)
```

```python
sns.barplot(class_distribution.index, class_distribution.values, alpha=0.8)
```

```python
X = dataset.loc[ : , "Alcohol":].copy()
y = dataset.loc[ : , "Class"].copy()
```

Before we start applying KMeans, we define a seed to be used for generating random numbers so that the results can be replicated every time we run the notebook.

```python
seed = 1234
```

We apply hierarchical clustering using single linkage. Given that there are 150 cases, we have 149 merges.

```python
def EvaluateKMeans(X, k_values, seed):
    wss_values = {}
    labels = {}

    for k in k_values:
        model = KMeans(n_clusters = k, random_state=seed)
        model.fit(X)
        labels[k] = model.predict(X)
        wss_values[k] = model.inertia_
    return wss_values, labels
```

```python
k_values = range(1,20)
wss_dict, labels = EvaluateKMeans(X, range(1,20),seed)
```

```python
wss_values = [wss_dict[x] for x in range(1,20)]
```

```python
fig = plt.figure(figsize=(8,6))
font = {'family' : 'sans', 'size'   : 16}
plt.rc('font', **font)
plt.plot(k_values, wss_values, 'bo-', color='red', label='WSS')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('WSS (Inertia)')
plt.xticks(k_values)
plt.legend()
plt.title('Hierarchical Clustering');
```

Let's zoom around 8, when the plot becomes flatter.

```python
fig = plt.figure(figsize=(8,6))
font = {'family' : 'sans', 'size'   : 16}
plt.rc('font', **font)
plt.plot(k_values[7:], wss_values[7:], 'bo-', color='red', label='WSS')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('WSS (Inertia)')
plt.xticks(k_values[7:])
plt.legend()
plt.title('Hierarchical Clustering');
```

We don't see much of an elbow until the very end. 

## External Meaures
Let's evaluate the solution with different numbers of clusters and compute a table that report the distribution of original labels in each one of the three clusters so that we can compare the clustering solution against the known labels.

```python
labels_single_k3 = labels[3]
df = pd.DataFrame({'Cluster': labels_single_k3, 'Class': dataset['Class']})

# Create crosstab: ct
ct = pd.crosstab(df['Cluster'],df['Class'])

# Print the table
print(ct)
```

```python
labels_single_k8 = labels[8]
df = pd.DataFrame({'Cluster': labels_single_k8, 'Class': dataset['Class']})

# Create crosstab: ct
ct = pd.crosstab(df['Cluster'],df['Class'])

# Print the table
print(ct)
```

As we can see only part of the data points belonging to class 1 are separated in cluster 1, 2, and 3. Elements of class 2 and 3 remain in cluster 2 mixed with elements of class 1.

## Clustering Normalized Data
k-Means works better if the variables have the similar means and variance. So we apply the standard scaler to the data and repeat the analysis.

```python
scaler = StandardScaler()
norm_X = scaler.fit_transform(X)
k_values = range(1,20)
wss_dict, norm_labels = EvaluateKMeans(norm_X, range(1,20),seed)
wss_values = [wss_dict[x] for x in range(1,20)]
fig = plt.figure(figsize=(8,6))
font = {'family' : 'sans', 'size'   : 16}
plt.rc('font', **font)
plt.plot(k_values, wss_values, 'bo-', color='red', label='WSS')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('WSS (Inertia)')
plt.xticks(k_values)
plt.legend()
plt.title('k-Means Clustering of Piedmont Wine Data');
```

We note that the WSS values are much lower since the variables have been normalized and their range is smaller. We also not an edge when k=3 and thus we check how the original labels map into the solution consisting of three clusters.

```python
labels_single_k3 = norm_labels[3]
df = pd.DataFrame({'Cluster': labels_single_k3, 'Class': dataset['Class']})

# Create crosstab: ct
ct = pd.crosstab(df['Cluster'],df['Class'])

# Print the table
print(ct)
```

As can be noted, we found three well-defined clusters identifying almost one class each.

## Clusters Summary
As the final step, let's summarize the clusters we found.

```python
clustering = X.copy()
clustering['Cluster'] = labels_single_k3
clustering.groupby('Cluster').mean()
```

Given the statistics, we should build a narrative to characterize the solution we selected. Such narrative is typically created using "cluster personas", profiles that describe the main features of the clusters.

## Snake Plots
Snake plots are market research techniques used to compare different segments. They provide a visual representation of each segment attributes. First, the data are normalized (according to the mean and the standard deviation) then we plot each cluster's average normalized values of each attribute.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

norm_dataset = norm_X.copy()
```

```python
norm_dataset = pd.DataFrame(norm_X,columns = dataset.columns[1:])
norm_dataset['Cluster'] = labels_single_k3
```

```python
melt_dataset = pd.melt(norm_dataset, 
                    id_vars=['Cluster'],
                    value_vars=dataset.columns[1:], 
                    var_name='Attribute', 
                    value_name='Value')
```

```python
import seaborn as sns
```

```python
plt.title('Snake plot of standardized variables')
sns.set(rc={'figure.figsize':(12,8)})
sns.set(font_scale = 1.5)
sns.lineplot(x="Attribute", y="Value", hue='Cluster', data=melt_dataset)
# g = sns.factorplot("year", data=planets, aspect=1.5, kind="count", color="b")
plt.xticks(rotation=90)
plt.xlabel('')
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1);
```
